```python
import json
from matplotlib import pyplot as plt
import numpy as np

with open('magnetic_data1.json', 'r') as j1:
    with open('magnetic_data2.json', 'r') as j2:
        data1 = json.load(j1)
        field1 = data1['field']
        distance1 = data1['distance']
        data2 = json.load(j2)
        field2 = data2['field']
        distance2 = data2['distance']
        # plt.legend()
        plt.plot(distance1,field1)
        plt.plot(distance2,field2)
        # np.array(field1)
```

```python
import dtw
for i in range(50):
    dtw_obj = dtw.dtw(field1[i:i+20], field2,step_pattern='asymmetric', open_begin=True, open_end=True,keep_internals=True)
    dtw_obj.plot('twoway',xlab=f'expected mapping {i}:{i+20}')
    # print()
```
